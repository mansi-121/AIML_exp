```python
import zipfile

# Open the zip file
with zipfile.ZipFile('business-employment-data-Jun-2024-quarter.zip', 'r') as z:

    # Load the specific CSV file from the zip into a pandas DataFrame
    with z.open('machine-readable-business-employment-data-Jun-2024-quarter.csv') as f:
        data = pd.read_csv(f)
```

Handling missing data

```python
# Fill missing values (example: filling with mean for numeric data)
data['Magnitude'].fillna(data['Magnitude'].mean(), inplace=True)

# Drop rows with missing values
data.dropna(inplace=True)
```

Correlation between numeric variables

```python
# Correlation matrix
correlation_matrix = data.corr()

# Display correlation matrix
correlation_matrix
```
